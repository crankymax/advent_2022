```python
## Day 2

d2_t = """A Y
B X
C Z
"""

#input_src = StringIO(d2_t)
input_src = open('d2_input.txt')

# A = Rock = X
# B = Paper = Y
# C = Scissors = Z

my_score_vals = {'X': 1, 'Y': 2, 'Z': 3}

score_val = { 'AX': 3, 'AY': 6, 'AZ': 0,
              'BX': 0, 'BY': 3, 'BZ': 6,
              'CX': 6, 'CY': 0, 'CZ': 3}
p1_total_score = 0

my_rsp_score = {'X': 0, 'Y': 3, 'Z': 6}

my_play = { 'AX': 'Z', 'AY': 'X', 'AZ': 'Y',
            'BX': 'X', 'BY': 'Y', 'BZ': 'Z',
            'CX': 'Y', 'CY': 'Z', 'CZ': 'X'}
p2_total_score = 0

for rsp_round in input_src:
    plays = rsp_round.split()
    p1_total_score += my_score_vals[plays[1]]+ score_val["".join(plays)]
    my_rsp= my_play["".join(plays)]
    p2_total_score += my_score_vals[my_rsp] + my_rsp_score[plays[1]]
    
print(p1_total_score)
print(p2_total_score)
```

```python
## Day 3
d3_t = """vJrwpWtwJgWrhcsFMMfFFhFp
jqHRNqRjqzjGDLGLrsFMfFZSrLrFZsSL
PmmdzqPrVvPwwTWBwg
wMqvLMZHhHMvwLHjbvcjnnSBnvTQFn
ttgJtRGJQctTZtZT
CrZsJsPPZsGzwwsLwLmpwMDw
"""

#input_src = StringIO(d3_t)
input_src = open('d3_input.txt')

priority = { x[1]: x[0]+1 for x in enumerate('abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ')}

score = 0
elf_count = 0
rucksacks = [0,0,0]
group_score = 0
for rucksack in input_src:
    elf_count += 1
    h_len = int(len(rucksack)/2)
    p1_s = set(rucksack[:h_len])
    p2_s = set(rucksack[h_len:-1])
    priority_f = p1_s.intersection(p2_s).pop()
    # print(priority_f, priority[priority_f])
    score += priority[priority_f]
    # Add rucksack to list pos
    rucksacks[(elf_count%3)-1] = set(rucksack[:-1])
    #print(rucksacks)
    #print(elf_count%3)
    if elf_count%3 == 0:
        common = rucksacks[0].intersection(rucksacks[1],rucksacks[2])
        # print(common)
        group_score += priority[common.pop()]

print(score)
print(group_score)
```

```python
## Day 7
d7_t = """$ cd /
$ ls
dir a
14848514 b.txt
8504156 c.dat
dir d
$ cd a
$ ls
dir e
29116 f
2557 g
62596 h.lst
$ cd e
$ ls
584 i
$ cd ..
$ cd ..
$ cd d
$ ls
4060174 j
8033020 d.log
5626152 d.ext
7214296 k
"""

#input_src = StringIO(d7_t)
input_src = open('d7_input.txt')

def parse(input_src):
    input_lines = input_src.read().split('\n')

    return input_lines

def build_tree(data):
    # nodes have name and size and children ['ll', 123, []]
    file_tree = ['/', 0, []]
    dir_stack = []
    cur_dir = file_tree
    for command in data:
        cm_pt = command.split(" ")
        if cm_pt[0] == '$' and cm_pt[1]=='cd':
            if '/' == cm_pt[2]:
                print(f'cd root')
                cur_dir = file_tree
                dir_stack = []
                continue
            elif cm_pt[2] == '..':
                # if no pop don't stop!
                if len(dir_stack) == 0:
                    print(f'Not Poppin {cur_dir[0]} with {cm_pt}')
                else:
                    # print(f'Poppin from {cur_dir[0]} to {"/".join([nm[0] for nm in dir_stack])}')
                    cur_dir = dir_stack.pop()
                continue
            
            fnd_dir = find_dir(cur_dir[2], cm_pt[2])

            if fnd_dir:
                dir_stack.append(cur_dir)
                cur_dir = fnd_dir
                # print(f'cd to {fnd_dir[0]} in {"/".join([nm[0] for nm in dir_stack])}')
            else:
                print(cm_pt)
                print(f'no find "{cm_pt[2]}" in {cur_dir[0]} ')
                show_tree(cur_dir)
                
        elif cm_pt[0].isnumeric():
            cur_dir[1] += int(cm_pt[0])
        elif cm_pt[0] == 'dir':
            cur_dir[2].append([cm_pt[1],0,[]])
        else:
            None
            # print(cm_pt)
    
    return file_tree
    
def find_dir(file_tree, rq_dir):
    for sd in file_tree:
        #print(f'srch={rq_dir} cur={sd}')
        if sd[0] == rq_dir:
            return sd
    
def add_subdir(tree):
    curr_levl = tree
    for dir_lvl in curr_levl[2]:
        tree[1] += add_subdir(dir_lvl)
    else:
        return tree[1]

def sum_subdir(tree, mx_sz):
    tot = 0
    for sd in tree:
        #print(sd)
        if sd[1] <= mx_sz:
            tot+=sd[1]
        if len(sd[2]) > 0:
            tot+=sum_subdir(sd[2],mx_sz)
    return tot

def show_tree(tree, prestr=''):
    for leaf in tree[2]:
        print(f':{prestr}/{leaf[0]}')
        if len(leaf[2])>0:
            show_tree(leaf, prestr+' ')
            
def dirsize_min(file_tree, req_disk, trg_dirs):
    for leaf in file_tree[2]:
        # print(leaf[0])
        if leaf[1] >= req_disk:
            trg_dirs.append((leaf[0],leaf[1]))
        if len(leaf[2])>0:
            dirsize_min(leaf, req_disk, trg_dirs)

def part_1(data):
    file_tree = build_tree(data)
    # add the dir sizes
    add_subdir(file_tree)
    
    tot_sum = sum_subdir(file_tree[2], 100000)
    
    print(tot_sum)
            
    
def part_2(data):
    tots_dsk = 70000000
    need_dsk = 30000000

    file_tree = build_tree(data)
    # add the dir sizes
    add_subdir(file_tree)
    
    unused = tots_dsk-file_tree[1]
    
    print(f'Unused = {unused}')
    req_disk = need_dsk - unused
    print(f'Required = {req_disk}')
    
    trgt_dirs = [(file_tree[0], file_tree[1])] 
    dirsize_min(file_tree, req_disk, trgt_dirs)
    
    #print(trgt_dirs)
    srtd_d = sorted(trgt_dirs, key= lambda tup: tup[1])
    print(f'Smallest dir is {srtd_d[0]}')    

data = parse(input_src)
print('Part 1')
part_1(data)

print('Part 2')
part_2(data)
```

```python
## Day 8
import numpy as np
d8_t = """30373
25512
65332
33549
35390"""

#input_src = StringIO(d8_t)
input_src = open('d8_input.txt')

def parse(input_src):
    input_lines = input_src.read().split('\n')
    grid = []
    for line in input_lines:
        if len(line.strip())==0:
            continue
        grid.append([t for t in line])
 
    # print(grid)

    np_grd = np.asmatrix(grid).astype(int)
    # print(np_grd.shape)    
    return np_grd

def part_1(data):
    # count the outside trees
    visible = 2*(data.shape[0] + data.shape[1]-2)
    
    print(f'Outside trees {visible}')
    for r in range(1, data.shape[0]-1):
        for c in range(1, data.shape[1]-1):
            #print(f'rc {r}{c} has to left max {np.max(data[r,:c])}')
            #print(f'rc {r}{c} has to right max {np.max(data[r,c+1:])}')
            #print(f'rc {r}{c} has to top max {np.max(data[:r,c])}')
            #print(f'rc {r}{c} has to bot max {np.max(data[r+1:,c])}')
            
            ct = data[r,c]
            
            if ct > np.max(data[r,:c]) or ct > np.max(data[r,c+1:]) or ct > np.max(data[:r,c]) or ct > np.max(data[r+1:,c]):
                visible += 1
            
    print(f'Visible {visible}')
    
def part_2(data):
    # print(data)
    mx_vw = 0
    for r in range(1, data.shape[0]-1):
        for c in range(1, data.shape[1]-1):
            t_hi = data[r,c]
            #print(f'Row {r} Col {c} height {t_hi}')
            # Add the left trees
            v_l = 0
            for lt in reversed(range(-c, 0)):
                #print(f'left {t_hi} > {data[r,c+lt]}  vl {v_l}')
                if data[r,c+lt] >= 0:
                    v_l += 1
                if t_hi <= data[r,c+lt]:
                    break

            # Add the right trees
            v_r = 0
            for rt in range(1,data.shape[1]-c):
                #print(f'right{t_hi} >  {data[r,c+rt]} vr {v_r}')
                if data[r,c+rt] >= 0:
                    v_r += 1
                if t_hi <= data[r,c+rt]:
                    break

            # Add the up trees
            v_u = 0
            for ut in reversed(range(-r, 0)):
                #print(f'up {t_hi} >{data[r+ut,c]}')
                if data[r+ut,c] >= 0:
                    v_u += 1
                if t_hi <= data[r+ut,c]:
                    break
            
            # Add the down trees
            v_d = 0
            for dt in range(1,data.shape[0]-r):
                #print(f'down {t_hi} > {data[r+dt,c]}')
                if data[r+dt,c] >= 0:
                    v_d += 1
                if t_hi <= data[r+dt,c]:
                    break

            v_t = v_l*v_r*v_u*v_d
            #print(f'Max view ({r},{c})= {v_t} = {v_l}*{v_r}*{v_u}*{v_d}')
            #print('')
            if mx_vw < v_t:
                mx_vw = v_t
    print(f'Max vw = {mx_vw}')

data = parse(input_src)
print('Part 1')
part_1(data)

print('\nPart 2')
part_2(data)
```
