```python
## Day 2

d2_t = """A Y
B X
C Z
"""

#input_src = StringIO(d2_t)
input_src = open('d2_input.txt')

# A = Rock = X
# B = Paper = Y
# C = Scissors = Z

my_score_vals = {'X': 1, 'Y': 2, 'Z': 3}

score_val = { 'AX': 3, 'AY': 6, 'AZ': 0,
              'BX': 0, 'BY': 3, 'BZ': 6,
              'CX': 6, 'CY': 0, 'CZ': 3}
p1_total_score = 0

my_rsp_score = {'X': 0, 'Y': 3, 'Z': 6}

my_play = { 'AX': 'Z', 'AY': 'X', 'AZ': 'Y',
            'BX': 'X', 'BY': 'Y', 'BZ': 'Z',
            'CX': 'Y', 'CY': 'Z', 'CZ': 'X'}
p2_total_score = 0

for rsp_round in input_src:
    plays = rsp_round.split()
    p1_total_score += my_score_vals[plays[1]]+ score_val["".join(plays)]
    my_rsp= my_play["".join(plays)]
    p2_total_score += my_score_vals[my_rsp] + my_rsp_score[plays[1]]
    
print(p1_total_score)
print(p2_total_score)
```

```python
## Day 3
d3_t = """vJrwpWtwJgWrhcsFMMfFFhFp
jqHRNqRjqzjGDLGLrsFMfFZSrLrFZsSL
PmmdzqPrVvPwwTWBwg
wMqvLMZHhHMvwLHjbvcjnnSBnvTQFn
ttgJtRGJQctTZtZT
CrZsJsPPZsGzwwsLwLmpwMDw
"""

#input_src = StringIO(d3_t)
input_src = open('d3_input.txt')

priority = { x[1]: x[0]+1 for x in enumerate('abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ')}

score = 0
elf_count = 0
rucksacks = [0,0,0]
group_score = 0
for rucksack in input_src:
    elf_count += 1
    h_len = int(len(rucksack)/2)
    p1_s = set(rucksack[:h_len])
    p2_s = set(rucksack[h_len:-1])
    priority_f = p1_s.intersection(p2_s).pop()
    # print(priority_f, priority[priority_f])
    score += priority[priority_f]
    # Add rucksack to list pos
    rucksacks[(elf_count%3)-1] = set(rucksack[:-1])
    #print(rucksacks)
    #print(elf_count%3)
    if elf_count%3 == 0:
        common = rucksacks[0].intersection(rucksacks[1],rucksacks[2])
        # print(common)
        group_score += priority[common.pop()]

print(score)
print(group_score)
```

```python
## Day 4
d4_t = """2-4,6-8
2-3,4-5
5-7,7-9
2-8,3-7
6-6,4-6
2-6,4-8
"""

#input_src = StringIO(d4_t)
input_src = open('d4_input.txt')

subs = 0
overs = 0
for rucksack in input_src:
    p1, p2 = rucksack.strip().split(',')
    r1 = p1.split('-')
    r2 = p2.split('-')
    s1 = set(range(int(r1[0]), int(r1[1])+1))
    s2 = set(range(int(r2[0]), int(r2[1])+1))
    # print(s1,s2)
    if s1.issubset(s2) or s2.issubset(s1):
        subs += 1
    
    if s1.intersection(s2) or s2.intersection(s1):
        overs +=1
        
print(subs)
print(overs)
```

```python
## Day 5
d5_t = """    [D]    
[N] [C]    
[Z] [M] [P]
 1   2   3 

move 1 from 2 to 1
move 3 from 1 to 3
move 2 from 2 to 1
move 1 from 1 to 2
"""

#input_src = StringIO(d5_t)
input_src = open('d5_input.txt')

input_lines = input_src.read().split('\n')

n_stack = int((int(len(input_lines[0])+1))/4)
print(n_stack)

stacks = [[] for d in range(n_stack)]

line_count = 0
for line in input_lines:
    if line.find('[')== -1:
        break

    line_count+=1
    for i in range(1, len(line), 4):
        #print(i//4, line[i:i+1])
        if line[i:i+1] != ' ':
            stacks[i//4].append(line[i:i+1])

for stk in stacks:
    print(stk)

for mv_com in input_lines[line_count+2:]:
    mp = mv_com.strip().split(' ')
    # print('Move command',mp)
    if len(mp) < 4:
        break
        
    nm = int(mp[1])
    fm = int(mp[3])-1
    dst = int(mp[5])-1
    # print('Target Stacks',stacks[fm][:nm])
    for crate in reversed(stacks[fm][:nm]):
        stacks[dst].insert(0, crate)
    del(stacks[fm][:nm])
        
    #print(stacks)


for stk in stacks:
    print(stk[0],end='')
print('')
```
